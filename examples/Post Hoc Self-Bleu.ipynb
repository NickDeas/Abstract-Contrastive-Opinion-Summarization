```python
import pandas as pd

from nltk.translate import bleu_score

import statistics
```

```python
def get_self_bleu(split_preds):
    self_bleus = []
    for pred_pair in split_preds:
        bleu   = bleu_score.sentence_bleu([pred_pair[0]], pred_pair[1], weights = (1.0,))
        self_bleus.append(bleu)
    return self_bleus
```

```python
def calc_file_self_bleu(file1, file2 = None):
    with open(file1, 'r', encoding = 'utf8') as f:
        preds = f.readlines()
        
    if file2:
        with open(file2, 'r', encoding = 'utf8') as f2:
            preds2 = f2.readlines()
            
    if file2:
        split_preds = list(zip(preds, preds2))
    else:
        split_preds = list(zip(preds[:len(preds)//2], preds[len(preds)//2:]))
        
    return get_self_bleu(split_preds)
```

```python
bart_fs_sbs    = calc_file_self_bleu('ex_summs_bart_fs.txt')
primera_fs_sbs = calc_file_self_bleu('ex_summs_primera_fs.txt')
```

```python
bart_sb = statistics.mean(bart_fs_sbs)
primera_sb = statistics.mean(primera_fs_sbs)
```

```python
print('Self-Bleu')
print(f'Bart FS: {bart_sb}')
print(f'Primera FS: {primera_sb}')
```
