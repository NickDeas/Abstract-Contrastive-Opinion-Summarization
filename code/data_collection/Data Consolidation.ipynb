```python
title_map = {news: title for title, newslist in zip(titles_news['title'].values, titles_news['first_news_real'].values) for news in newslist}
```

```python
title_map = {(key[:key.find('?')] if '?' in key else key) :val for key, val in title_map.items()}
```

```python
def get_orig_title(row):
    url_cols = [f'urls/{i}/expanded_url' for i in range(10)]
    row_urls = row[url_cols].dropna().values
    row_urls = [(url[:url.find('?')] if '?' in url else url) for url in row_urls]
    
    for url in row_urls:
        if url in title_map.keys():
            return title_map[url]
    
    return None
```

```python
tweets['title'] = tweets.progress_apply(get_orig_title, axis = 1)
```

# Dataset Statistics 

```python
flipside_num = pd.read_csv('../../data/data_clean/flipside_w_meta.csv')
all_sm_groups = pd.read_csv('../../data/data_clean/all_source_concat.csv')
```

```python
flipside_num.head(2)
```

```python
flipside_ref = flipside_num[~flipside_num['left_sum'].isna()]
reddit_texts = all_sm_groups[all_sm_groups['sm_src'] == 'reddit']
tweet_texts  = all_sm_groups[all_sm_groups['sm_src'] == 'twitter']
all_texts    = all_sm_groups.groupby('src_title')['text'].apply(lambda x: '||'.join(x)).reset_index()

reddit_texts['splits']  = reddit_texts["text"].str.split(r"[\.\!\?\\\-\|]")
reddit_texts['splits']  = reddit_texts["splits"].apply(lambda l: [t for t in l if len(t) > 5])
reddit_texts['num_sents'] = reddit_texts['splits'].apply(len)
reddit_texts['avg_len'] = reddit_texts['splits'].apply(lambda l: sum([len(sent.split(' ')) for sent in l])/len(l))

tweet_texts['splits']  = tweet_texts["text"].str.split(r"[\.\!\?\\\-\|]")
tweet_texts['splits']  = tweet_texts["splits"].apply(lambda l: [t for t in l if len(t) > 5])
tweet_texts['num_sents'] = tweet_texts['splits'].apply(len)
tweet_texts['avg_len'] = tweet_texts['splits'].apply(lambda l: sum([len(sent.split(' ')) for sent in l])/len(l))

all_texts['splits']  = all_texts["text"].str.split(r"[\.\!\?\\\-\|]")
all_texts['splits']  = all_texts["splits"].apply(lambda l: [t for t in l if len(t) > 5])
all_texts['num_sents'] = all_texts['splits'].apply(len)
all_texts['avg_len'] = all_texts['splits'].apply(lambda l: sum([len(sent.split(' ')) for sent in l])/len(l))
```

```python
print(f'Total Articles: {len(all_sm_groups["src_title"].unique())}')
print(f'# References:   {len(flipside_ref)}')
print(f'Avg. LS Len:    {flipside_ref["left_sum"].str.len().mean():.3f}')
print(f'Avg. RS Len:    {flipside_ref["right_sum"].str.len().mean():.3f}')
print(f'Avg. Num R:     {reddit_texts["num_sents"].mean():.3f}')
print(f'Avg. Num T:     {tweet_texts["num_sents"].mean():.3f}')
print(f'Avg. Len R:     {reddit_texts["avg_len"].mean():.3f}')
print(f'Avg. Len T:     {tweet_texts["avg_len"].mean():.3f}')
print(f'Avg. Num Sent: {all_texts["num_sents"].mean(): .3f}')
print(f'Avg. Len Total: {all_texts["avg_len"].mean():.3f}')
```
